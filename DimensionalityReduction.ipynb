# PCA

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load iris dataset
data = load_iris()
X = data.data
y = data.target

# Plotting the non-linearly separable dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('First Feature')
plt.ylabel('Second Feature')
plt.title('t')
plt.colorbar(label='Target')
plt.show()

# Center the data (optional since PCA in sklearn does this automatically)
# X_centered = X - np.mean(X, axis=0)

# Perform PCA
# pca = PCA(n_components=2)
# X_pca = pca.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Explained variance ratio
explained_variance = pca.explained_variance_ratio_

print(f"Explained variance ratio for the 2 components: {explained_variance}")

# Plotting the data projected onto the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - Iris dataset')
plt.colorbar(label='Target')
plt.show()
```

# Kernel PCA for Non-Linear Data`

```python
from sklearn.decomposition import KernelPCA
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

# Generate a non-linearly separable dataset (e.g., moons)
X, y = make_moons(n_samples=100, noise=0.1, random_state=42)

# Plotting the non-linearly separable dataset
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('First Feature')
plt.ylabel('Second Feature')
plt.title('Moons dataset')
plt.colorbar(label='Target')
plt.show()

# # Perform Kernel PCA (with RBF kernel)
kpca = KernelPCA(n_components=2 , kernel= 'rbf' , gamma = 15)
X_kpca = kpca.fit_transform(X)
# kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
# X_kpca = kpca.fit_transform(X)

# Plotting the transformed data
plt.figure(figsize=(8, 6))
plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=y, cmap='viridis')
plt.xlabel('Kernel Principal Component 1')
plt.ylabel('Kernel Principal Component 2')
plt.title('Kernel PCA - Moons dataset')
plt.colorbar(label='Target')
plt.show()
```
